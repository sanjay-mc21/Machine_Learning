```python
import pandas as pd
df1=pd.read_csv('titanic_1.csv')
df1
```

```python
df=df1.drop(columns = ['PassengerId' , 'Name' , 'SibSp' , 'Parch' , 'Cabin' , 'Ticket' , 'Embarked'])
df
```

```python
dum = pd.get_dummies(df['Sex'] , dtype = 'int')
dum
```

```python
dum
```

```python
df
```

```python
df = pd.concat([df,dum],axis = 'columns')
df
```

```python
df.drop(columns=['Sex' , 'female'] , inplace = True)
df
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df
```

```python
df.isna().sum()
```

```python
x=df.drop(columns = ['Survived'])
y=df['Survived']
x
```

```python
from sklearn.model_selection import train_test_split as tts
x_train , x_test , y_train , y_test = tts(x,y,test_size = 0.3)
```

```python
from sklearn.naive_bayes import GaussianNB as gnb
mod = gnb()
mod
```

```python
mod1=gnb()
mod1
```

```python
mod.fit(x_train , y_train)
```

```python
mod1.fit(x_test , y_test)
```

```python
mod.score(x_train , y_train)
```

```python
mod1.score(x_test , y_test)
```

```python
mod.predict(x_test)
```

```python
mod.predict_proba(x_test)
```
