```python
from sklearn.datasets import load_iris as l_i
iris = l_i()
iris
```

```python
import pandas as pd
df=pd.DataFrame(iris['data'],columns=iris['feature_names'])
df
```

```python
df['target'] = iris['target']
df
```

```python
df['flower name'] = df['target'].apply(lambda x : iris['target_names'][x])
df
```

```python
x=df.drop(columns=['sepal width (cm)','sepal length (cm)','flower name','target'])
y=df['flower name']
y
x
```

```python
from sklearn.model_selection import train_test_split as tts
x_train,x_test,y_train,y_test=tts(x,y,test_size=0.2)
```

```python
from sklearn.svm import SVC as svc
mod = svc()
mod
```

```python
mod.fit(x_train,y_train)
```

```python
mod.score(x_test,y_test)
```

```python
mod1 = svc(C=10)
mod1
```

```python
mod1.fit(x_train,y_train)
```

```python
mod1.score(x_test,y_test)
```

```python
mod2 = svc(gamma=1)
mod2
```

```python
mod2.fit(x_train,y_train)
```

```python
mod2.score(x_test,y_test)
```
